```python
range_total_ESG_risk = ESG['Total ESG Risk score'].max() - ESG['Total ESG Risk score'].min()
print(f"The range of 'Total ESG Risk score' is: {range_total_ESG_risk}")
```

```python
import numpy as np
```

```python
ESG = ESG.dropna()
```

```python
quantile_ESG = (ESG['Total ESG Risk score'], .27)
```

```python
np.quantile(ESG['Total ESG Risk score'], .27)
```

```python
q1, q3 = np.quantile(ESG['Total ESG Risk score'],[.25, .75])
manual_iqr = q3-q1
f"The IQR of no_siblings by manual compuation is {manual_iqr}"
```

```python
from scipy.stats import iqr
```

```python
scipy_iqr = iqr(ESG['Total ESG Risk score'])  # Use the correct iqr function
print(f"The IQR of Total ESG Risk score using the scipy function is {scipy_iqr}")

# Calculating lower and upper fences
lower_fence = np.quantile(ESG['Total ESG Risk score'], 0.25) - 1.5 * scipy_iqr
upper_fence = np.quantile(ESG['Total ESG Risk score'], 0.75) + 1.5 * scipy_iqr

print(f"Lower fence is {lower_fence}")
print(f"Upper fence is {upper_fence}")

# Boxplot for visualization
import matplotlib.pyplot as plt

plt.boxplot(ESG['Total ESG Risk score'])
plt.title("Boxplot of Total ESG Risk Score")
plt.ylabel("Score")
plt.show()
```

```python
ESG.loc[ : , 'Total ESG Risk score'].hist(bins=30)
plt.title('Histogram for Total ESG Risk, bins = 30')
plt.show()
```

```python
import pandas as pd

# Assuming 'ESG' is your DataFrame (Ensure it is loaded before using)
# ESG = pd.read_csv("your_file.csv")  # Uncomment this to load your dataset

# Convert numeric columns from string to float, coercing errors to NaN
ESG = ESG.apply(pd.to_numeric, errors='coerce')

# Drop non-numeric columns to avoid correlation errors
numeric_ESG = ESG.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numeric_ESG.corr()

# Calculate the basic statistics using describe
describe_stats = numeric_ESG.describe()

# Concatenate the correlation matrix and describe statistics
result = pd.concat([describe_stats, correlation_matrix], keys=["Statistics", "Correlation"])

# Print the result
print(result)
```
